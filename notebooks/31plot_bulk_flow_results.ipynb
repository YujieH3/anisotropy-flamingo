```python
import matplotlib.pyplot as plt
plt.style.use('./nice.mplstyle')
import pandas as pd
data = pd.read_csv('../data/fits/bulk_flow_lightcone0_all.csv')
```

```python
d = data.loc[(data['scaling_relation']=='LX-T') & (data['zmax']==0.17)]
d.sort_values('sigma')
```

```python
fig, axes = plt.subplots(ncols=3, nrows=3, figsize=(10, 10), sharex=True, sharey=True)

# Adjust the spacing between plots
plt.subplots_adjust(wspace=0, hspace=0)

zmaxs = data['zmax'].unique()
i = 0
for row in axes:
    ylabel=True
    for ax in row:
        d = data.loc[(data['scaling_relation']=='LX-T') & (data['zmax']==zmaxs[i])]
        ax.scatter(d['ubf'], d['sigma'], s=3)
        ax.set_xlabel('Bulk flow')
        if ylabel:
            ax.set_ylabel('Scatter')
            ylabel = False
        ax.set_title(f'zmax={zmaxs[i]}', fontsize=15, y=0.88)

        i += 1
        if i >= len(zmaxs):
            break
    if i >= len(zmaxs):
        break

plt.show()
```
