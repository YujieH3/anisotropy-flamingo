```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
# plt.style.use('nice.mplstyle')

data = pd.read_csv('./data/fits/bootstrap_LX-T.csv')
bins = np.linspace(data['A'].min(), data['A'].max(), 11)
data['A'].hist(bins=10, alpha=0.5, density=True, label='1000')
data['A'][:500].hist(bins=bins, alpha=0.5, density=True, label='500') # density
# data['A'][:200].hist(bins=bins, alpha=0.5, density=True, label='200')
# data['A'][:100].hist(bins=bins, alpha=0.5, density=True, label='100')
plt.legend()
plt.show()
```

```python
# percentile
print(np.percentile(data['A'], [16, 84]))
print(np.percentile(data['A'][:500], [16, 84]))
print(np.percentile(data['A'][:150], [16, 84]))
```

```python
x = np.arange(10, 1000, 10)
y1 = [np.percentile(data['A'][:N], 16) for N in range(10, 1000, 10)]
y2 = [np.percentile(data['A'][:N], 84) for N in range(10, 1000, 10)]
plt.plot(x, y1, label='16th percentile')
plt.plot(x, y2, label='84th percentile')
plt.legend()
plt.show()
```
