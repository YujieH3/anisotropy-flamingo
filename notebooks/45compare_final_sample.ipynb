**1. Basic info**

```python
import pandas as pd
import numpy as np
mock = pd.read_csv('../data/mock_lightcone/sample_lightcone/samples_in_lightcone0000.csv')
mock
```

```python
real = pd.read_csv('../data/samples_in_lightcone0_with_trees_duplicate_excision.csv')
real
```

Redshift distribution

```python
import matplotlib.pyplot as plt
plt.style.use('./nice.mplstyle')
plt.hist(real['redshift'].values, histtype='step', range=(0, 0.35), label='true')
plt.hist(mock['redshift'].values, histtype='step', range=(0, 0.35), label='mock')
plt.legend()
plt.xlabel('$z_\\mathrm{{cos}}$')
plt.ylabel('Count')
plt.show()
```

**2. Halo matching**

```python
# Extract the overlaps
true_soapid = real['SOAPID'].values
true_snapnum = real['snap_num'].values

mock_soapid = mock['SOAPID'].values
mock_snapnum = mock['snap_num'].values

# the halos from mock lightcones that are in the halo lightcones
mock_overlap_mask = np.isin(mock_soapid, true_soapid) & np.isin(mock_snapnum, true_snapnum)
mock_overlap = mock[mock_overlap_mask]

# the halos from halo_lightcones that are in our mock samples
true_overlap_mask = np.isin(true_soapid, mock_soapid) & np.isin(true_snapnum, mock_snapnum)
true_overlap = real[true_overlap_mask]

print(len(mock_overlap), len(true_overlap))

mock_overlap.sort_values(by='redshift', inplace=True)
mock_overlap = mock_overlap.drop_duplicates(subset=['SOAPID', 'snap_num'], keep='first') 

true_overlap.sort_values(by='redshift', inplace=True)
true_overlap = true_overlap.drop_duplicates(subset=['SOAPID', 'snap_num'], keep='first')

mock_overlap.sort_values(by=['snap_num', 'SOAPID'], inplace=True)
true_overlap.sort_values(by=['snap_num', 'SOAPID'], inplace=True)
print(len(mock_overlap), len(true_overlap))
```
