```python
import pandas as pd
import numpy as np
import sys
sys.path.append('../tools')
import clusterfit as cf

# Read in the bulk flow data
i=0

data = pd.read_csv(f'../data/samples_in_lightcone{i}_with_trees_duplicate_excision_outlier_excision.csv')
mcmc_file = f'../data/fits/bulk_flow_mcmc_lightcone{i}.csv'
best_fit_file = f'../data/fits/bulk_flow_lightcone{i}.csv'
bootstrap_file = f'../data/fits/bulk_flow_bootstrap_lightcone{i}.csv'

name = ['\\LXT', '\\YSZT', '\\MgasT']
for j, relation in enumerate(['LX-T', 'YSZ-T', 'M-T']):
    n_clusters = cf.CONST[relation]['N']
    true_zmaxs, true_ubfs, true_vlons, true_vlats = cf.true_bulk_flow_z(data, zrange0=0.06, zrange1=0.13, zstep=0.02, n_clusters=n_clusters) # ubf_z returns a list
    inversed_vlons, inversed_vlats = cf.opposite_direction_arr(true_vlons, true_vlats)

    zmaxs, mc_ubfs, mc_ubf_lowers, mc_ubf_uppers, mc_vlons, mc_vlon_lowers, mc_vlon_uppers, mc_vlats, mc_vlat_lowers, mc_vlat_uppers = cf.read_bulk_flow_mcmc(mcmc_file, relation=relation, radian=False)
        
    # # Get best fit parameters
    # zmaxs, ubfs, vlons, vlats = cf.read_bulk_flow(best_fit_file, relation=relation, radian=False)

    # Get bootstapping constrain 
    zmaxs, de_ubfs, de_ubf_lowers, de_ubf_uppers, de_vlons, de_vlon_lowers, de_vlon_uppers, de_vlats, de_vlat_lowers, de_vlat_uppers = cf.read_bulk_flow_bootstrap(
        bootstrap_file=bootstrap_file, best_fit_file=best_fit_file, 
        relation=relation, radian=False, median=True)
    
    for k, zmax in enumerate(zmaxs):

        if k == 0:
            _name_ = name[j]
        else:
            _name_ = ''

        zmax = zmaxs[k]
        mask = (data[:n_clusters]['ObservedRedshift'] < zmax)
        num = np.sum(mask)

        print(f'{_name_} & $z<{zmaxs[k]:.2f}$ & {num}')

        print(f' & ${true_ubfs[k]:.2f}$ & $(\\ang{{{inversed_vlons[k]:.0f}}}, \\ang{{{inversed_vlats[k]:.0f}}})$')
        
        print(f'& ${de_ubfs[k]:.0f}^{{+{de_ubf_uppers[k]:.0f}}}_{{-{de_ubf_lowers[k]:.0f}}}$ &',
            f'$(\\ang{{{de_vlons[k]:.0f}}}^{{+\\ang{{{de_vlon_uppers[k]:.0f}}}}}_{{-\\ang{{{de_vlon_lowers[k]:.0f}}}}},',
            f'\\ang{{{de_vlats[k]:.0f}}}^{{+\\ang{{{de_vlat_uppers[k]:.0f}}}}}_{{-\\ang{{{de_vlat_lowers[k]:.0f}}}}})$')

        print(f' & ${mc_ubfs[k]:.0f}^{{+{mc_ubf_uppers[k]:.0f}}}_{{-{mc_ubf_lowers[k]:.0f}}}$ &',
            f'$(\\ang{{{mc_vlons[k]:.0f}}}^{{+\\ang{{{mc_vlon_uppers[k]:.0f}}}}}_{{-\\ang{{{mc_vlon_lowers[k]:.0f}}}}},',
            f'\\ang{{{mc_vlats[k]:.0f}}}^{{+\\ang{{{mc_vlat_uppers[k]:.0f}}}}}_{{-\\ang{{{mc_vlat_lowers[k]:.0f}}}}})$') 
        

        print('\\\\')

    print('\n\\midrule\n')
```

```python

import pandas as pd
import numpy as np
data = pd.read_csv(f'../data/fits/bulk_flow_mcmc_lightcone1.csv')
for i in range(len(data)):
    slice = data.iloc[i]
    zmax = slice['zmax']

    ubf = slice['ubf']
    ubf_l = slice['ubf_err_lower']
    ubf_u = slice['ubf_err_upper']

    lon = slice['vlon']
    lon_l = slice['vlon_err_lower']
    lon_u = slice['vlat_err_upper']

    lat = slice['vlat']
    lat_l = slice['vlon_err_lower']
    lat_u = slice['vlat_err_upper']

    ubf = round(ubf)
    ubf_l = round(ubf_l)
    ubf_u = round(ubf_u)
    lat = round(lat)
    lat_l = round(lat_l)
    lat_u = round(lat_u)
    lon = round(lon)
    lon_l = round(lon_l)
    lon_u = round(lon_u)

    relation = slice['scaling_relation']
    if relation == 'LX-T':
        name = '\\LXT'
    elif relation == 'YSZ-T':
        name = '\\YSZT'
    elif relation == 'M-T':
        name = '\\MgasT'

    print(f'{name} & $z<{zmax:.2f}$ & ${ubf}^{{+{ubf_u}}}_{{-{ubf_l}}}$ &',
          f'$(\\ang{{{lon}}}^{{+\\ang{{{lon_u}}}}}_{{-\\ang{{{lon_l}}}}},',
          f'   \\ang{{{lat}}}^{{+\\ang{{{lat_u}}}}}_{{-\\ang{{{lat_l}}}}})$ \\\\')
```
