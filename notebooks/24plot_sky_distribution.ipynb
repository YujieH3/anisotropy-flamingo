```python
import pandas as pd
data = pd.read_csv('/data1/yujiehe/data/samples_in_lightcone0_with_trees_duplicate_excision_outlier_excision.csv')
print(len(data))
data = data[:313]
```

Sky distribution of our final sample


```python
import matplotlib.pyplot as plt
import numpy as np
plt.style.use("./nice.mplstyle")
fig = plt.figure(figsize=(15, 7))
ax = plt.subplot(111, projection="hammer")
ax.grid(True)
# plt.axis('off')

l = data["phi_on_lc"] * np.pi / 180
b = data["theta_on_lc"] * np.pi / 180

p = ax.scatter(l, b, c=data['ObservedRedshift'], cmap="Reds") # takes (longitude, latitude)
# plt.scatter([0, np.pi/4, 0], [0, 0, np.pi/4], marker='+', c='r')

plt.colorbar(p, label="observed redshift", ax=ax)
plt.savefig('../data/plots/sky-map.png', dpi=300, transparent=True)
plt.show()
```
