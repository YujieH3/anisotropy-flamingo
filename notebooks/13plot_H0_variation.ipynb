```python
import matplotlib.pyplot as plt
plt.style.use("./nice.mplstyle")
import pandas as pd
import numpy as np
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap

df = pd.read_csv('./data/fits/scan_best_fit_LX-T_Î¸75.csv')

fig = plt.figure(figsize=(10, 4))
m = Basemap(projection='hammer', lon_0=0)

data = np.array(df['A']).reshape(90, 90).T

lon = np.linspace(-180, 180, 90)
lat = np.linspace(-90, 90, 90)
lons, lats = np.meshgrid(lon, lat)
x, y = m(lons, lats)

# Display the data as an image; set the extent of the map
im = m.imshow(70*np.sqrt(data/1.514), extent=[-180, 180, -90, 90], cmap='inferno')

plt.colorbar(im, label='$H_0$ (km/s/Mpc)', )
plt.title('$L_\\mathrm{{X}}-T\\quad\\theta=75\\degree$', loc='right')
plt.show()
```
