````python
import pandas as pd
pd.set_option('display.max_columns', None)      # Display all columns when printing
original_data = pd.read_hdf("/data1/yujiehe/data/halo_properties_in_lightcone0.hdf5", "lightcone")
````

````python
# Drop duplicate
print(len(original_data))
original_data.sort_values(by="redshift", inplace=True)
data = original_data.drop_duplicates(subset=["M_fof_lc"], keep="first", inplace=False)
data = data[data.SpectroscopicLikeTemperatureWithoutRecentAGNHeatingCoreExcision > 0]    # there are infs
data = data[data.M500 > 5e12]
print(len(data))
````
