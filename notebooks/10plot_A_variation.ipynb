# One plot

```python
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap

plt.style.use("./nice.mplstyle")
import pandas as pd
import numpy as np

fig, ax = plt.subplots(3, 2, figsize=(20, 15))
plt.subplots_adjust(hspace=0.1, wspace=0.05)

m = Basemap(projection='hammer', lon_0=0)

lon = np.linspace(-180, 180, 90)
lat = np.linspace(-90, 90, 90)
lons, lats = np.meshgrid(lon, lat)
x, y = m(lons, lats)


files = ['../data/fits/scan_best_fit_LX-T_75.csv', 
         '../data/fits/scan_best_fit_YSZ-T_60.csv',
         '../data/fits/scan_best_fit_M-T_75.csv', 
         '../data/fits/scan_best_fit_LX-YSZ_75.csv',
         '../data/fits/scan_best_fit_LX-M_75.csv',
         '../data/fits/scan_best_fit_YSZ-M_75.csv',
        ]

titles = ['$L_\\mathrm{{X}}-T\\quad\\theta=75\\degree$',
          '$Y_\\mathrm{{SZ}}-T\\quad\\theta=60\\degree$',
          '$M_\\mathrm{{gas}}-T\\quad\\theta=75\\degree$',
          '$L_\\mathrm{{X}}-Y_\\mathrm{{SZ}}\\quad\\theta=75\\degree$',
          '$L_\\mathrm{{X}}-M_\\mathrm{{gas}}\\quad\\theta=75\\degree$',
          '$Y_\\mathrm{{SZ}}-M_\\mathrm{{gas}}\\quad\\theta=75\\degree$',
        ]

best_fit_As = [1.514, 1.079, 1.069, 2.972, 1.230, 3.829]

idx = 0
for row in ax:
    for subax in row:
        subax.set_xlabel('Longitude')
        subax.set_ylabel('Latitude')
        # Load the data from the file
        df = pd.read_csv(files[idx])
        data = np.array(df['A']).reshape(90, 90).T
        # Display the data as an image; set the extent of the map
        im = m.imshow(data/best_fit_As[idx], extent=[-180, 180, -90, 90], 
                      cmap='tab20b', ax=subax,
                      vmin=0.90, vmax=1.15)

        plt.colorbar(im, label='$A/A_\\mathrm{{all}}$', ax=subax)
        subax.set_title(titles[idx], loc='right')
        idx += 1

plt.show()      
```
