```python
import h5py

def h5ls(f):
    if type(f) == h5py._hl.dataset.Dataset:
        print(f, f[:10])
    else:
        print(f.keys())
        for key in f.keys():
            h5ls(f[key])

with h5py.File('../data/halo_properties_in_lightcones.hdf5', 'r') as f:
    h5ls(f)
```
