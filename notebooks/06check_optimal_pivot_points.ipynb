```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('nice.mplstyle')
data = pd.read_csv('./data/tests/pivot_bootstrap_LX-T_4.csv')
plt.scatter(data['A'], data['B'], marker='x')
plt.xlabel('$A$')
plt.ylabel('$B$')
plt.title('$L_\\mathrm{{X}}-T\\quadC_X=4$')
plt.grid(True)
plt.show()
```
