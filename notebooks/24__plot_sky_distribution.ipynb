Sky distribution of our final sample


```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

for lc in [0, 1]:

    data = pd.read_csv(f'../data/samples_in_lightcone{lc}_with_trees_duplicate_excision_outlier_excision.csv')
    print(len(data))
    data = data[:313]
    plt.style.use("./nice.mplstyle")
    fig = plt.figure(figsize=(12, 5))
    ax = plt.subplot(111, projection="hammer")
    ax.grid(True)
    # plt.axis('off')

    l = data["phi_on_lc"] * np.pi / 180
    b = data["theta_on_lc"] * np.pi / 180

    p = ax.scatter(l, b, c=data['ObservedRedshift'], cmap="Reds",
                   vmin=0, vmax=0.20) # takes (longitude, latitude)
    # plt.scatter([0, np.pi/4, 0], [0, 0, np.pi/4], marker='+', c='r')

    print(np.min(data['ObservedRedshift']), np.max(data['ObservedRedshift']))
    print(np.median(data['ObservedRedshift']))
    print(np.mean(data['ObservedRedshift']))
    plt.colorbar(p, label="Observed Redshift", ax=ax)
    plt.savefig(f'/Users/yujiehe/Documents/projects/project-anisotropy/plots/forpaper/sample-sky-distribution-lc{lc}.pdf', bbox_inches='tight')
    plt.show()
```
