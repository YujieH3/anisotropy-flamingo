### Default M21 method, default temperature

```python
import clusterfit as cf
import os
import pandas as pd
import numpy as np
```

```python
data_dir = '/cosma8/data/do012/dc-he4/analysis'
all_data = []

for lc in range(1728):
    lc00 = f'{lc:04d}'
    file = f'{data_dir}/lc{lc00}/fit_all_scatter.csv'
    cat_file = f'{data_dir}/lc{lc00}/samples_in_lightcone{lc00}_duplicate_excision_outlier_excision_scatter_assigned.csv'
    if os.path.exists(file):
        df = pd.read_csv(file)
        cat = pd.read_csv(cat_file)

        df['TotalScat'] = 0
        # Calculate total scatter
        for relation in ['LX-T', 'YSZ-T']:
            yname, xname = cf.parse_relation_name(relation)
            N = cf.CONST[relation]['N']    # number of cluster

            eY = cat['e'+cf.COLUMNS[yname]][:N].values  # sigma_logY
            eX = cat['e'+cf.COLUMNS[xname]][:N].values  # sigma_logX
            scat_Y = np.log10(1 + eY) 
            scat_X = np.log10(1 + eX)

            scat_int = df[df['Relation'] == relation]['BestFitScat'].values # sigma_intr
            B = df[df['Relation'] == relation]['BestFitB'].values # B

            scat_tot = np.sqrt(scat_Y**2 + B**2 * scat_X**2 + scat_int**2)
            df['TotalScat'][df['Relation'] == relation] = np.median(scat_tot)

        # Label lightcone number
        df['Lightcone'] = lc
        
        all_data.append(df)

# Concatenate all dataframes
df = pd.concat(all_data, ignore_index=True)
```

```python
for relation in ['LX-T', 'YSZ-T']:
    mask = df['Relation'] == relation
    masked_df = df[mask]
    print(relation)
    for key in ['BestFitA', 'BestFitB', 'BestFitScat', 'TotalScat']:
        arr = masked_df[key].values
        median = np.median(arr)
        eightyfourth = np.nanpercentile(arr, 84)
        sixteenth = np.nanpercentile(arr, 16)
        print(f'{key} = {median:.3f}^{{+{eightyfourth - median:.3f}}}_{{-{median - sixteenth:.3f}}}')
```

### MCMC temperature

```python
data_dir = '/cosma8/data/do012/dc-he4/analysis'
all_data = []

for lc in range(1728):
    lc00 = f'{lc:04d}'
    flag = f'{data_dir}/lc{lc00}/fit-all-mc-scatter.done'
    file = f'{data_dir}/lc{lc00}/fit_all_mc_scatter.csv'
    cat_file = f'{data_dir}/lc{lc00}/samples_in_lightcone{lc00}_duplicate_excision_outlier_excision_scatter_assigned_mc.csv'
    if os.path.exists(file) and os.path.exists(flag):
        df = pd.read_csv(file)
        cat = pd.read_csv(cat_file)

        df['TotalScat'] = 0
        # Calculate total scatter
        for relation in ['LX-T', 'YSZ-T']:
            yname, xname = cf.parse_relation_name(relation)
            N = cf.CONST_MC[relation]['N']    # number of cluster

            eY = cat['e'+cf.COLUMNS_MC[yname]][:N].values  # sigma_logY
            eX = cat['e'+cf.COLUMNS_MC[xname]][:N].values  # sigma_logX
            scat_Y = np.log10(1 + eY) 
            scat_X = np.log10(1 + eX)

            scat_int = df[df['Relation'] == relation]['BestFitScat'].values # sigma_intr
            B = df[df['Relation'] == relation]['BestFitB'].values # B

            scat_tot = np.sqrt(scat_Y**2 + B**2 * scat_X**2 + scat_int**2)
            df['TotalScat'][df['Relation'] == relation] = np.median(scat_tot)

        # Label lightcone number
        df['Lightcone'] = lc
        
        all_data.append(df)

# Concatenate all dataframes
df = pd.concat(all_data, ignore_index=True)
```

```python
for relation in ['LX-T', 'YSZ-T']:
    mask = df['Relation'] == relation
    masked_df = df[mask]
    print(relation)
    for key in ['BestFitA', 'BestFitB', 'BestFitScat', 'TotalScat']:
        arr = masked_df[key].values
        median = np.median(arr)
        eightyfourth = np.nanpercentile(arr, 84)
        sixteenth = np.nanpercentile(arr, 16)
        print(f'{key} = {median:.3f}^{{+{eightyfourth - median:.3f}}}_{{-{median - sixteenth:.3f}}}')
```
