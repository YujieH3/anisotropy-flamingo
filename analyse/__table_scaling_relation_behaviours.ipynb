# Plot the full sky best fit behaviours
The temperature used is the Chandra temperature. As default by the M21 analysis.

```python
import os
import pandas as pd
import numpy as np

data_dir = '/cosma8/data/do012/dc-he4/analysis'
all_data = []

for lc in range(1728):
    lc00 = f'{lc:04d}'
    filename = f'{data_dir}/lc{lc00}/fit_all.csv'
    if os.path.exists(filename):
        df = pd.read_csv(filename)
        df['Lightcone'] = lc
        all_data.append(df)

# Concatenate all dataframes
df = pd.concat(all_data, ignore_index=True)
```

```python
# Write the concatenated dataframe to a CSV file
output_filename = f'/cosma8/data/do012/dc-he4/analysis_all/concatenated_fit_all.csv'
df.to_csv(output_filename, index=False)
```

```python
import pandas as pd
df = pd.read_csv('/cosma8/data/do012/dc-he4/analysis_all/concatenated_fit_all.csv')
df
```

```python
df.drop_duplicates()
```

```python
for relation in ['LX-T', 'YSZ-T', 'M-T']:
    mask = df['Relation'] == relation
    masked_df = df[mask]
    print(relation)
    for key in ['BestFitA', 'BestFitB', 'BestFitScat']:
        arr = masked_df[key].values
        median = np.median(arr)
        eightyfourth = np.percentile(arr, 84)
        sixteenth = np.percentile(arr, 16)
        print(f'{key} = {median:.3f} +{eightyfourth - median:.3f} -{median - sixteenth:.3f}')
```
