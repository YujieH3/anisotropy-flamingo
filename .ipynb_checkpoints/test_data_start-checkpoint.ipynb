```python
# SOAP = '/net/hypernova/data2/FLAMINGO/L2800N5040/HYDRO_FIDUCIAL/SOAP/halo_properties_0038.hdf5'
SOAP = '/net/hypernova/data2/FLAMINGO/L1000N1800/HYDRO_FIDUCIAL/SOAP/halo_properties_0077.hdf5'
# SOAP = '/net/hypernova/data2/FLAMINGO/FLAMINI/L1000N0084/HYDRO_FIDUCIAL/SOAP/SOAP_properties_0077.hdf5'
with h5py.File(SOAP) as f:
    print(f["SO/500_crit"].keys())
    # SOAP_M500 = f["SO/500_crit/TotalMass"][:]
    # SOAP_CY = f["SO/5xR_500_crit/ComptonY"][:]
    # SOAP_T = f["SO/500_crit/GasTemperature"][:]
    SOAP_TCE = f["SO/500_crit/GasTemperatureCoreExcision"][:]
    # SOAP_XR0 = f["SO/500_crit/XRayLuminosity"][:,0]
    # SOAP_XR1 = f["SO/500_crit/XRayLuminosity"][:,1]
    SOAP_XR2 = f["SO/500_crit/XRayLuminosity"][:,0]
```
